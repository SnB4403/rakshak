```python
## stitching images

import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
wp1 = cv2.imread('1.jpg')
wp2 = cv2.imread('2.jpg')
wp3 = cv2.imread('3.jpg')
wp4 = cv2.imread('4.jpg')
wp5 = cv2.imread('5.jpg')
wp6 = cv2.imread('6.jpg')
wp7 = cv2.imread('7.jpg')
wp8 = cv2.imread('8.jpg')
```

```python
# stitch 8 images together

plt.figure(figsize=(10,10))
plt.subplot(2,4,1)
plt.imshow(cv2.cvtColor(wp1, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.subplot(2,4,2)
plt.imshow(cv2.cvtColor(wp2, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.subplot(2,4,3)
plt.imshow(cv2.cvtColor(wp3, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.subplot(2,4,4)
plt.imshow(cv2.cvtColor(wp4, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.subplot(2,4,5)
plt.imshow(cv2.cvtColor(wp5, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.subplot(2,4,6)
plt.imshow(cv2.cvtColor(wp6, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.subplot(2,4,7)
plt.imshow(cv2.cvtColor(wp7, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.subplot(2,4,8)
plt.imshow(cv2.cvtColor(wp8, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.subplots_adjust(wspace=None, hspace=None)
plt.show()
```
